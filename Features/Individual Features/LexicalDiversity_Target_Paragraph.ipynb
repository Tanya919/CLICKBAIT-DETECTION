```python
tok = ld.tokenize(paragraph[0])
print(tok[:40])
```

```python
l = []
for para in paragraph:
  if type(para)==float:
    l.append(0)
    continue
  flt = ld.flemmatize(para)
  #print(flt)
  lexD= ld.ttr(flt)
  #print(lexD)
  l.append(round(lexD,2))

 

  

  

```

```python
print(l)
```

```python
df["lexical Diversity"] = l
```
