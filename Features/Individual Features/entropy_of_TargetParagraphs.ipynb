```python
def remove_stopwords(str_list):
  stop_words = set(stopwords.words('english'))
  filtered_sentence = [w.lower() for w in str_list if not w.lower() in stop_words]
  return filtered_sentence
    
```

```python
def entropyUtil(sen):
   PreprossedSen = sen.translate(str.maketrans('', '', string.punctuation))
   dictionary = {}
   wordCount = []
   lst = PreprossedSen.split()
   removedStopWord = remove_stopwords(lst)
   for elements in removedStopWord:
     if elements[-1] == '.':
       elements = elements[0:len(elements) - 1]
     if elements in dictionary:
       dictionary[elements] += 1
     else:
       dictionary.update({elements: 1})
   for allKeys in dictionary:
       wordCount.append(dictionary[allKeys])
   count = len(removedStopWord)
  #  print(count)
   ent = 0
   for freq in wordCount:
      prob = freq/count
      # print(prob)
      ent += -(prob * log2(prob)) 
   return ent
   
```

```python
def entropy(para):
  l=[]
  for item in para:
    if len(para)==0:
      l.append(0)
      continue
    en = entropyUtil(item)
    l.append(en)
  return l
```

```python
entropy_l = entropy(paragraph)
```
