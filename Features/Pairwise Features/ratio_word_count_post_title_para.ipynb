```python
import re
def remove_emoji(string):
    emoji_pattern = re.compile("["
                           u"\U0001F600-\U0001F64F" # emoticons
                           u"\U0001F300-\U0001F5FF" # symbols & pictographs
                           u"\U0001F680-\U0001F6FF" # transport & map symbols
                           u"\U0001F1E0-\U0001F1FF" # flags (iOS)
                           u"\U00002702-\U000027B0"
                           u"\U000024C2-\U0001F251"
                           "]+", flags=re.UNICODE)
    return emoji_pattern.sub(r'', string)

for i in range(len(postText)):
  t = remove_emoji(postText[i])
  postText[i] = t
```

```python
def count_word(para, text, title):
  l_para, l_text, l_title = [], [], []
  for i in range(len(para)):
    s1 = re.sub(r'[^\w\s]', '', para[i])
    tokens_para = nltk.word_tokenize(s1)
    s2 = re.sub(r'[^\w\s]', '', text[i][2:-2])
    tokens_text = nltk.word_tokenize(s2)
    s3 = re.sub(r'[^\w\s]', '', title[i])
    tokens_title = nltk.word_tokenize(s3)
    l_para.append(len(tokens_para))
    l_text.append(len(tokens_text))
    l_title.append(len(tokens_title))
  return l_para, l_text, l_title
```

```python
l_para, l_text, l_title = count_word(paragraph,postText,title)
```

```python
def ratio_para_postText(para,text):
  l = []
  for i in range(len(para)):
    if para[i] == 0 or text[i]==0:
      print(para[i])
      l.append(0)
    else:
      l.append(text[i]/para[i])
  return l
```

```python
ratioParaPosttext = ratio_para_postText(l_para, l_text)
```

```python
ratioTitlePosttext = ratio_text_Title(l_text,l_title)
```
